```python
# File: app.py
from flask import Flask, request, jsonify
from transformers import BertTokenizer, TFBertModel
import torch
# from pyngrok import ngrok  # Add this line to integrate Ngrok
from pyngrok import ngrok  # Add this line to integrate Ngrok

# Load the Hugging Face model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")

app = Flask(__name__)

# Prediction endpoint
@app.route("/predict", methods=["POST"])
def predict():
    data = request.json
    input_text = data.get("text", "")

    # Tokenize input
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True)

    # Make prediction
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.softmax(outputs.logits, dim=-1)

    # Get the highest scoring label
    label = torch.argmax(predictions, dim=1).item()
    confidence = predictions[0][label].item()

    return jsonify({
        "label": label,
        "confidence": confidence
    })

if __name__ == "__main__":
    # Open an Ngrok tunnel on port 5000
    ngrok.set_auth_token("2lz6l2eT9NzCuVjJKb0XRZgh9tb_4whbezf8YYSXEuci9LXwM")
    public_url = ngrok.connect(5000)
    print(f" * Ngrok tunnel opened at {public_url}")



    # Start the Flask app
    app.run(port=5000)
```
